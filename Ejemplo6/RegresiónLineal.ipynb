## REGRESIÓN LINEAL - OUTLIERS

# 1) primero cargamos las librerias 

```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# 2) vamos a crear un conjunto de datos base para trabajar

```python
np.random.seed(0)
X = np.random.normal(0, 1, 100) #datos de entrada
y =  2 * X + 1 + np.random.normal(0, 0.5, 100) #datos de salida
print("datos de entrada X:", X[-5:])
print("datos de salida y:", y[-5:])
```

# 3) A manera de ejemplo, vamos a agregar outliers, este es un proceso de descubrimiento imaginemos que son parte de los datos originales

```python
X_outliers = np.append(X, [3, 3.5, 5])
y_outliers = np.append(y, [10, 20, 15])
print("datos de entrada con outliers X:", X_outliers[-5:])
print("datos de salida con outliers y:", y_outliers[-5:])
X_outliers = X_outliers.reshape(-1, 1) # este paso es necesario para que sklearn pueda trabajar con los datos 
#lo que hace es convertir el vector en una matriz de una sola columna
print("datos de entrada con outliers X:", X_outliers[-5:])

```

# 4) vamos a utilizar la herramienta LinearRegression que nos ofrece la libreria de scikit-learn

```python
model = LinearRegression() # Creamos el modelo de regresión lineal
model.fit(X_outliers, y_outliers) # Entrenamos el modelo con los datos de entrada y salida con outliers
y_pred = model.predict(X_outliers) # Predecimos los valores de salida con el modelo entrenado
```

# 5) vamos a visualizar el resultado 

```python

plt.figure(figsize=(8, 5))
plt.scatter(X_outliers, y_outliers, color='blue', label='Datos (con outliers)')
plt.plot(X_outliers, y_pred, color='red', label='Regresión lineal')
plt.title("Regresión Lineal con Outliers")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(X_outliers, y_outliers, color='blue', label='Datos (con outliers)')
plt.plot(X_outliers, y_pred, color='red', label='Regresión lineal')
plt.scatter([3, 3.5, 5], [10, 20, 15], color='green', label='Outliers', s=100) # puntos de outliers
plt.title("Regresión Lineal con Outliers")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
